# 500 articles for topic sports

```python
from nytimesarticle import articleAPI
from bs4 import BeautifulSoup  
from urllib.request import urlopen
#import urllib2
import csv
import time
```

```python
#https://pypi.org/project/NYTimesArticleAPInew/
#https://stackoverflow.com/questions/40090962/extracting-articles-from-new-york-post-by-using-python-and-new-york-post-api
api = articleAPI("wETCvjx1EQ8MroxZdMboRhA3DJ0v4TEX")
links=[]
for i in range(0,50):
    article = api.search( q = 'sports', page=i, begin_date = 20190404, end_date=20190411)
    time.sleep(10)
    #print(article['response'],['docs'])
    for i in article['response']['docs']:
            links.append(i['web_url'])  
#print(links)
```

```python
#https://docs.python.org/2/howto/urllib2.html
#https://stackoverflow.com/questions/2792650/import-error-no-module-name-urllib2
import string
data_content = []
finalstripped=[]

for data in links:
    html=urlopen(data)
    r=html.read()
    time.sleep(2)
    #https://www.youtube.com/watch?v=zXif_9RVadI
    soup = BeautifulSoup(r)
    soupval = soup.findAll('p')
    finaltext = ""
    for values in soupval:
        #https://stackoverflow.com/questions/43675635/adding-a-space-between-paragraphs-when-extracting-text-with-beautifulsoup
        currenttext=values.getText(separator="\n", strip=True)
        finaltext = finaltext +"\n"+ currenttext
    #print(finaltext)
    data_content.append(finaltext)
    
#print(data_content)
```

```python
#https://stackoverflow.com/questions/16923281/pandas-writing-dataframe-to-csv-file
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_csv.html
import pandas as pd
df = pd.DataFrame(data_content)
df.to_csv('nytimesdata.txt',sep='\t', index=False)
```
