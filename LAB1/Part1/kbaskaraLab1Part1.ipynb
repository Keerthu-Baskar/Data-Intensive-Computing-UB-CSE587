# Problem 1
 - In this program the two sales datas are plotted in a graph. For sales1 the entire data is taken, sales2 gives a poisson distribution of 12 numbers with mean as 34.
 - Sales1 graph is fixed whereas sales2 graph changes in every run because of the poisson distribution.
 ## Analysis from plot
 - From the plot we can make out that sales1 has more variations and fluctuations in the value and sales2 has more fluctuations.
 - The plot of sales1 ranges roughly between 17 and 43. which means the median is 30. And the mean of sales1 is around 22
 - The mean of sales2 is 34 and the median is lies around 35.
 - We can see that sales2 value is greater than sales1

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
# rpois Generates random variates of 12 values which has mean 34
sales2<-rpois(12,34)
#par is the graphical parameter. We are getting the background color of graph here
par(bg="cornsilk")
#the plotting function used in R is plot(). We will pass a vector and we will get a scatter plot of magnitude vs index
#generally the parameters are X,Y and type of plot
#but in certain cases the parameter Y can be optional if X is appropriate structure
plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
#title of the plot
title(main="Sales by Month")
#The lines() function adds information to a graph. 
#It can not produce a graph on its own. Usually it follows a plot(x, y) command that produces a graph.
#IT is a function taking coordinates and joins the points with line segments.
lines(sales2, type="o", pch=22, lty=2, col="red")
#adds a rectangular grid to the plot
grid(nx=NA, ny=NULL)
#scale given in top right box
#legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)

```

# Problem 3
 - This graph clearly shows the range and median of sales1 and sales2.
 - The range and median of sales2 is greater than sales1.
 ## Analysis from plot
 - The box or whisker plot creates a visual distribution of the quantitive data. It splits the data as percentile. 
 - The bottom line of box is 25% percentile and the middle dark line is the median. 
 - The top line is 75% percentile. 
 - The outer line are the whiskers. They are the lowest and highest value of data. 
 - All other plots that may appear above or below the whiskers are the outliers in the data.
 - Generally value greater than 1.5 interquartile ranges away from the 25th or 75th percentile
 - Boxplot tells the central value and the variability in the data (range)
 - Boxplot also gives skew of the data. (where most of the data lies)

```r
#stats a vector of length 5, containing the extreme of the lower whisker, the lower ‘hinge’, the median, the upper ‘hinge’ and the extreme of the upper whisker.
#$stats are sorted in increasing order
fn<-boxplot(sales,col=c("orange","green"))$stats
#adding text to a plot
text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```
